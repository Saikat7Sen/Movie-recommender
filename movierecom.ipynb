```python
import numpy as np
import pandas as pd
```

```python
credits_data = pd.read_csv("/content/credits.csv")
movies_data = pd.read_csv("/content/movies.csv")
```

```python
movies_data
```

```python
movies_data = movies_data.merge(credits_data, on = "title")
```

```python
movies_data.head()
```

```python
movies_data = movies_data[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
```

```python
movies_data.isnull().sum()
```

```python
movies_data.dropna(inplace = True)
```

```python
movies_data.duplicated().sum()
```

```python
movies_data.iloc[0].genres
```

```python
import ast
```

```python
def convert (obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
```

```python
movies_data["genres"] = movies_data["genres"].apply(convert)
movies_data["keywords"] = movies_data["keywords"].apply(convert)
movies_data.head()
```

```python
def convert3 (obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i["name"])
      counter += 1
    else:
      break
    return L
```

```python
movies_data["cast"] = movies_data["cast"].apply(convert3)
movies_data.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      L.append(i["name"])
      break
  return L
```

```python
movies_data["crew"] = movies_data["crew"].apply(fetch_director)
movies_data.head()
```

```python
movies_data["overview"] = movies_data["overview"].apply(lambda x:x.split())
```

```python
movies_data.head()
```

```python
movies = movies_data.copy()
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x] if x is not None else [])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x] if x is not None else [])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x] if x is not None else []) # Handle None values
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x] if x is not None else []) # Handle None values
movies
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"] + movies["overview"]
```

```python
movies
```

```python
new_df = movies[["movie_id", "title", "tags"]]
```

```python
new_df
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
cv.fit_transform(new_df["tags"]).toarray().shape
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```

```python
recommend("Titanic")
```
